```python
import os
import cv2
from cvzone.HandTrackingModule import HandDetector
from cvzone.ClassificationModule import Classifier
import numpy as np
import math
import time
import tensorflow as tf
```

```python
#Initialize camera
cap = cv2.VideoCapture(0)

# Get the current working directory
base_dir = os.getcwd()

# Define paths to the model and labels files
model_path = os.path.join(base_dir, "Model/keras_model.h5")
labels_path = os.path.join(base_dir, "Model/labels.txt")

# Check if the files exist
if not os.path.exists(model_path):
    raise FileNotFoundError(f"Model file not found: {model_path}")
if not os.path.exists(labels_path):
    raise FileNotFoundError(f"Labels file not found: {labels_path}")

# Initialize hand detector and classifier
detector = HandDetector(maxHands=1)
classifier = Classifier(model_path, labels_path)

# Define constants
offset = 20
imgSize = 400

# Define folder for data and counter
folder = 'Data/C'
counter = 0

# Define labels
labels = ["A", "B", "C"]

# Example image processing function
def process_image(image_path):
    img = cv2.imread(image_path)
    imgOutput = img.copy()
    hands, img = detector.findHands(img)

    if hands:
        hand = hands[0]
        x, y, w, h = hand['bbox']
        imgWhite = np.ones((imgSize, imgSize, 3), np.uint8) * 255
        imgCrop = img[y - offset:y + h + offset, x - offset:x + w + offset]
        imgCropShape = imgCrop.shape
        aspectRatio = h / w

        if aspectRatio > 1:
            k = imgSize / h
            wCal = math.ceil(k * w)
            imgResize = cv2.resize(imgCrop, (wCal, imgSize))
            imgResizeShape = imgResize.shape
            wGap = math.ceil((imgSize - wCal) / 2)
            imgWhite[:, wGap:wCal + wGap] = imgResize
        else:
            k = imgSize / w
            hCal = math.ceil(k * h)
            imgResize = cv2.resize(imgCrop, (imgSize, hCal))
            imgResizeShape = imgResize.shape
            hGap = math.ceil((imgSize - hCal) / 2)
            imgWhite[hGap:hCal + hGap, :] = imgResize

        prediction, index = classifier.getPrediction(imgWhite, draw=False)
        print(f"Prediction: {prediction}, Index: {index}")
        if 0 <= index < len(labels):
            cv2.putText(imgOutput, labels[index], (x, y - 20), cv2.FONT_HERSHEY_COMPLEX, 2, (255, 0, 255), 2)
            print("Result is ", labels[index])
        else:
            print(f"Invalid index: {index}, length of labels: {len(labels)}")
            cv2.putText(imgOutput, "Unknown", (x, y - 20), cv2.FONT_HERSHEY_COMPLEX, 2, (255, 0, 255), 2)

        return imgOutput
    else:
        print("No hands detected")
        return img

# Example usage
image_path = "path/to/your/image.jpg"
processed_image = process_image(image_path)
```
